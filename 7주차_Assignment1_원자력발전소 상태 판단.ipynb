# **<선형판별분석 (LDA)>**

**: 원자력 발전소 상태 판단**

* **데이터셋 설명**
  - 한국수력원자력(주)에서 제공받은 원자력 발전소 모사 데이터

* **데이터파일 설명**

  - **pca_train.csv** : 훈련 데이터가 포함된 csv 파일
  - **pca_test.csv** : 테스트 데이터가 포함된 csv 파일
  - **train_label.csv** : 훈련데이터의 id (파일 이름)에 해당되는 상태_B를 정리한 파일
  - **sample_submission.csv** : 제출 형식과 동일한 샘플 데이터



```python
import pandas as pd
import numpy as np
```

```python
train_x = pd.read_csv('/content/pca_train.csv')
train_y = pd.read_csv('/content/train_label.csv')

test_x = pd.read_csv('/content/pca_test.csv')

submit = pd.read_csv('/content/sample_submission.csv')
```

# 데이터 살펴보기

```python
train_x
```

```python
train_y = train_y['label']
train_y
```

```python
test_x
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
```

# 데이터 전처리

```python
#표준화
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(train_x)
train_x_std = scale.transform(train_x)
test_x_std = scale.transform(test_x)
```

# 모델

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

cld = LinearDiscriminantAnalysis(store_covariance=True, n_components=2)       #n_components 확인
#cqd = QuadraticDiscriminantAnalysis(store_covariance=True)

cld.fit(train_x_std, train_y)
#cqd.fit(train_x_std, train_y)
```

# 예측

```python
test_y_pred = cld.predict(test_x_std)
#test_y_pred = cqd.predict(test_x_std)
test_y_pred
```

# 저장

```python
for i in range(len(test_y_pred)) :
  submit['label'][i] = test_y_pred[i]

submit
```

```python
submit.to_csv('원자력발전소_18012576.csv', index=False)
```
