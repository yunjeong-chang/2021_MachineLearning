```python
#데이터 불러오기
import pandas as pd
train = pd.read_csv('/content/data.csv')
test_x = pd.read_csv('/content/X_test.csv')
submit = pd.read_csv('/content/sample_submission.csv')
```

```python
#타겟 변수 분리
train_x = train.drop(['TARGET'], axis=1)
train_y = train['TARGET']
```

```python
train_x
```

```python
train_x.describe()
```

```python
#필요한 변수만 정규화
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
train_x['T'] = scale.fit_transform(train_x.loc[:,train_x.columns=='T'])
test_x['T'] = scale.fit_transform(test_x.loc[:,test_x.columns=='T'])

train_x['RH'] = scale.fit_transform(train_x.loc[:,train_x.columns=='RH'])
test_x['RH'] = scale.fit_transform(test_x.loc[:,test_x.columns=='RH'])
```

```python
#회귀나무 모델
from sklearn.tree import DecisionTreeRegressor
dtr1 = DecisionTreeRegressor(max_depth = 12, random_state=1) #12 : 99635
dtr1.fit(train_x, train_y)
```

```python
# 예측 및 저장
test_y_pred1 = dtr1.predict(test_x)

submit['TARGET'] = test_y_pred1
submit
```

```python
#제출 파일 생성
submit.to_csv('submit_18012576.csv', index=False)
```
